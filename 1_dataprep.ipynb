# Data Pre-processing

```python
# data cleaning tracker 
report = pd.DataFrame({'task': 'start',
                       'nrow_changed': np.nan,
                       'ncol_changed': np.nan ,
                       'nrow_after': df0.shape[0],
                       'ncol_after': df0.shape[1]}, index = [0])
report
```

### Filter age groups

```python
df1 = df0[df0['AGEG10LFS_T'].isin([2,3,4])]
# df1 = df0.query("AGEG10LFS_T in (2,3,4)")
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'filter age groups',
                                     'nrow_changed': (df1.shape[0] - report.loc[max(report.index), 'nrow_after']), 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```

### Remove NaNs in select cols

```python
df1 = df1.dropna(subset=['YEARLYINCPR','AGEG10LFS_T','RACETHN_4CAT','EDCAT6','GENDER_R','J_Q04A','REGION_US'])
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'remove NaNs',
                                     'nrow_changed': (df1.shape[0] - report.loc[max(report.index), 'nrow_after']), 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```

### Create new variables or recode them into the same ones

```python
# Low and High income
df1['LowIncome'] = np.where(df1['YEARLYINCPR'].isin([1,2]), 1, 0)
df1['HighIncome'] = np.where(df1['YEARLYINCPR'].isin([5,6]), 1, 0)
df1['YEARLYINCPR_R'] = np.where(df1['YEARLYINCPR'].isin([1,2]), 'Q1', 
                                np.where(df1['YEARLYINCPR']==3, 'Q2',
                                         np.where(df1['YEARLYINCPR']==4, 'Q3', 'Q4')))
df1['LowIncome'].value_counts(dropna=False)
df1['HighIncome'].value_counts(dropna=False)
df1['YEARLYINCPR_R'].value_counts(dropna=False)
```

```python
# recode age
df1['AGEG10LFS_T'] = df1['AGEG10LFS_T'].replace({2: '2534', 3: '3544', 4: '4554'})
df1['AGEG10LFS_T'].value_counts(dropna=False)
```

```python
# recode race
df1['RACETHN_4CAT'] = df1['RACETHN_4CAT'].replace({1: 'Hispanic', 2: 'White', 3: 'Black', 6: 'Other'})
df1['RACETHN_4CAT'].value_counts(dropna=False)
```

```python
# recode edcat into 3 groups
df1['EDCAT3'] = np.where(df1['EDCAT6'].isin([1,2]), 'Secondary',
                         np.where(df1['EDCAT6']==3, 'Non-tertiary', 'Tertiary'))
df1['EDCAT3'].value_counts(dropna=False)
```

```python
# recode Gender and create FEMALE
df1['FEMALE'] = np.where(df1['GENDER_R']==2, 1, 0)
df1['GENDER_R'] = np.where(df1['GENDER_R']==1, 'Male', 'Female')
df1['FEMALE'].value_counts(dropna=False)
df1['GENDER_R'].value_counts(dropna=False)
```

```python
# recode born in the country
df1['US_born'] = np.where(df1['J_Q04A']==1, 1, 0)
df1['US_born'].value_counts(dropna=False)
```

```python
# recode US regions
df1['REGION_US'] = np.where(df1['REGION_US']==1, 'Northeast',
                            np.where(df1['REGION_US']==2, 'Midwest',
                                     np.where(df1['REGION_US']==3, 'South', 'West')))
df1['REGION_US'].value_counts(dropna=False)
```

```python
# calculate scaled sum of pvlit and pvnum as cog_skills
## df1['STLITNUM1'] = np.asarray(df1[['PVLIT1', 'PVNUM1']].sum(axis=1))
## df1['STLITNUM1'] = (df1['STLITNUM1'] - df1['STLITNUM1'].mean()) / df1['STLITNUM1'].std()
from sklearn.preprocessing import scale
lit_cols = df1.filter(regex='^PVLIT').columns
num_cols = df1.filter(regex='^PVNUM').columns
for i in range(len(lit_cols)):
    col_name = 'STLITNUM{}'.format(i + 1)
    df1[col_name] = scale(df1[[lit_cols[i], num_cols[i]]].sum(axis=1).values.reshape(-1, 1))
# Note that scale requires the input to be a 2-dimensional array, so 
# we use the values.reshape(-1, 1) code to ensure that the input is a 2-dimensional array
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'create new cols',
                                     'nrow_changed': (df1.shape[0] - report.loc[max(report.index), 'nrow_after']), 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```

### Remove outliers

```python
grouped = df1.groupby(['GENDER_R', 'RACETHN_4CAT', 'EDCAT3'])
df1['q1'] = grouped['EARNMTHBONUSUS_C'].transform('quantile', 0.25)
df1['q3'] = grouped['EARNMTHBONUSUS_C'].transform('quantile', 0.75)
df1['iqr'] = df1['q3'] - df1['q1']
df1['lower'] = df1['q1'] - 4 * df1['iqr']
df1['upper'] = df1['q3'] + 4 * df1['iqr']
df1['EARNMTHBONUSUS_C_outlier'] = np.where(df1['EARNMTHBONUSUS_C'] > df1['upper'], 'high',
                                           np.where(df1['EARNMTHBONUSUS_C'] < df1['lower'], 'low',
                                                    np.where(df1['EARNMTHBONUSUS_C'].isna(), 'missing', 'normal')))
df1 = df1[df1['EARNMTHBONUSUS_C_outlier'] != 'high']
df1 = df1.drop(columns=['q1', 'q3', 'iqr', 'lower', 'upper', 'EARNMTHBONUSUS_C_outlier'])
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'remove outliers',
                                     'nrow_changed': (df1.shape[0] - report.loc[max(report.index), 'nrow_after']), 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```

### Merge groups

```python
# Merge groups: Other-Non-tertiary cell is relatively small; recode them as Other-Tertiary
# save number of rows affected
nrow_changed = df1[(df1['RACETHN_4CAT']=="Other") & (df1['EDCAT3']=="Non-tertiary")].shape[0]
# merge
df1['EDCAT3'] = np.where((df1['RACETHN_4CAT']=="Other") & (df1['EDCAT3']=="Non-tertiary"), "Tertiary", df1['EDCAT3'].astype(str))
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'merge groups',
                                     'nrow_changed': "(" + str(nrow_changed) + ")", 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```

### Replace NaNs

```python
# replace NA EARNMTHBONUSUS_C with the mean incomes for YEARLYINCPR
# save number of rows affected
nrow_changed = df1['EARNMTHBONUSUS_C'].isna().sum()
# replace NA 
grouped = df1.groupby('YEARLYINCPR')
df1['EARNMTHBONUSUS_C_mean'] = grouped['EARNMTHBONUSUS_C'].transform('mean')
df1['EARNMTHBONUSUS_C'] = np.where(df1['EARNMTHBONUSUS_C'].isna(), df1['EARNMTHBONUSUS_C_mean'], df1['EARNMTHBONUSUS_C'])
df1 = df1.drop(columns = ['EARNMTHBONUSUS_C_mean'])
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'replace NaNs',
                                     'nrow_changed': "(" + str(nrow_changed) + ")", 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```

### Select columns

```python
cols = (['LowIncome','HighIncome','FEMALE',
        'C_Q07','YEARLYINCPR','GENDER_R','AGEG10LFS_T','RACETHN_4CAT','EDCAT6','EDCAT3',
        'C_Q09_C','ISCOSKIL4','EARNMTHBONUSUS_C','EARNMTHBONUSPPPUS_C','D_Q18A_T',
        'J_Q04A','US_born','REGION_US'] + 
        list(df1.loc[:, 'LITSTATUS':'PVNUM10'].columns) + 
        list(df1.loc[:, 'STLITNUM1':'STLITNUM10'].columns) +
        ['VEMETHOD','VENREPS'] + list(df1.loc[:, 'SPFWT0':'SPFWT80'].columns))
df1 = df1.loc[:, cols]
```

```python
# update the tracker 
report = report.append(pd.DataFrame({'task': 'select cols',
                                     'nrow_changed': (df1.shape[0] - report.loc[max(report.index), 'nrow_after']), 
                                     'ncol_changed': (df1.shape[1] - report.loc[max(report.index), 'ncol_after']), 
                                     'nrow_after': df1.shape[0], 
                                     'ncol_after': df1.shape[1]}, index = [max(report.index)+1]))
report
```
